```python
import pandas as pd

# open all the files in the data directory as dataframes

tickers = ['AAPL', 'AMZN', 'DIA', 'EEM', 'META', 'MSFT', 'QQQ', 'SPY', 'TSLA', 'VXX']

frames = []

for e in tickers: 
    filename = '../data/' + e + '_1min_firstratedata.csv'
    df = pd.read_csv(filename)
    df['ticker'] = e
    frames.append(df)

# show the correlation between the closing prices of each of the stocks

df = pd.concat(frames)
```

```python
# Show correlations between the closing prices of each of the stocks

df = df[['ticker', 'close']]
df = df.pivot(columns='ticker', values='close')
df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(), annot=True)
```
